```python
import pandas as pd
import konlpy
```

```python
kkma=konlpy.tag.Kkma()
okt=konlpy.tag.Okt()
han=konlpy.tag.Hannanum()
```

## 1. Merge data

```python
df1 = pd.read_csv('Book_90020001.csv')
df2 = pd.read_csv('Book_90040001.csv')
df3 = pd.read_csv('Book_90060000.csv')
```

```python
df = pd.concat([df1, df2, df3], ignore_index = True)
```

```python
df
```

```python
df.to_csv("38235.csv")
```

## 2. KoNLPy

```python
story = df['Story']
```

```python
index = story.index
```

```python
list = []
for i in index:
    list.append(story[i])
list
```

```python
list[0]
```

```python
okt.pos(str(list[0]))
```

```python
story_pos = []
for i in range(len(story)):
    story_pos.append(okt.pos(str(list[i])))
```

```python
story_pos[0]
```

## 3. Noun

```python
nouns = []
for i in range(len(story_pos)):
    list = []
    for j in story_pos[i]:
        if j[1] == "Noun":
            list.append(j[0])
    nouns.append(list)
```

```python
nouns
```

## 4. Adjective

```python
adjs = []
for i in range(len(story_pos)):
    list = []
    for j in story_pos[i]:
        if j[1] == "Adjective":
            list.append(j[0])
    adjs.append(list)
```

```python
adjs
```

## 5. To dataframe

```python
df['Adj'] = 0
for i in range(len(df)):
    df['Adj'][i] = str(' '.join(adjs[i]))
```

```python
for i in range(len(df)):
    df['Noun'][i] = str(' '.join(nouns[i]))
        
```

```python
df['NounAdj'] = 0
for i in range(len(df)):
    df['NounAdj'][i] = df['Noun'][i] + df['Adj'][i]
```
